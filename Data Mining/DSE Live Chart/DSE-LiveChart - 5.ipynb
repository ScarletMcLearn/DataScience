```python
from bokeh.plotting import figure, show
from bokeh.models import Legend, LegendItem, ColumnDataSource, Circle, Line

def create_plot(x_values, y_values, plot_type='circle', line_color='blue', line_width=2, circle_size=10, circle_fill_color='blue', circle_border_color='black', circle_border_width=2, p=None):
    # Create a figure and a data source if not provided
    if p is None:
        p = figure()

    # Create a legend item for the plot
    legend_item = LegendItem(label=plot_type.capitalize())

    if plot_type == 'circle':
        # Add circles to the figure
        circle_source = ColumnDataSource(data=dict(x=x_values, y=y_values))
        circle_renderer = p.circle(x='x', y='y', size=circle_size, fill_color=circle_fill_color, line_color=circle_border_color, line_width=circle_border_width, source=circle_source)
        legend_item.renderers = [circle_renderer]
    elif plot_type == 'line':
        # Add a line to the figure
        line_renderer = p.line(x=x_values, y=y_values, line_color=line_color, line_width=line_width)
        legend_item.renderers = [line_renderer]

    # Create the legend and add the legend item
    legend = Legend(items=[legend_item])

    # Add the legend to the figure
    p.add_layout(legend)

    # Return the figure
    return p
```

```python
from bokeh.plotting import figure, show
from bokeh.models import Legend, LegendItem, ColumnDataSource

def create_plot(x_values, y_values, plot_type='circle', line_color='blue', line_width=2, shape_size=10, shape_fill_color='blue', shape_border_color='black', shape_border_width=2, p=None):
    if p is None:
        # Create a new figure if p is not provided
        p = figure()

    if plot_type == 'circle':
        # Add circles to the figure
        shape_source = ColumnDataSource(data=dict(x=x_values, y=y_values))
        shape_renderer = p.circle(x='x', y='y', size=shape_size, fill_color=shape_fill_color, line_color=shape_border_color, line_width=shape_border_width, source=shape_source)
        return p, shape_renderer
    elif plot_type == 'diamond':
        # Add diamonds to the figure
        shape_source = ColumnDataSource(data=dict(x=x_values, y=y_values))
        shape_renderer = p.diamond(x='x', y='y', size=shape_size, fill_color=shape_fill_color, line_color=shape_border_color, line_width=shape_border_width, source=shape_source)
        return p, shape_renderer
    elif plot_type == 'line':
        # Add a line to the figure
        line_source = ColumnDataSource(data=dict(x=x_values, y=y_values))
        line_renderer = p.line(x='x', y='y', line_color=line_color, line_width=line_width, source=line_source)
        return p, line_renderer

    return p, None

# Create a plot with shapes and lines
x_1_values = [1, 2, 3, 4, 5]
y_1_values = [2, 4, 6, 8, 10]
x_2_values = [6, 7, 8, 9, 10]
y_2_values = [11, 12, 13, 14, 15]

p = figure()

p, shape_renderer1 = create_plot(x_1_values, y_1_values, plot_type='circle', shape_fill_color='blue', p=p)
p, line_renderer1 = create_plot(x_1_values, y_1_values, plot_type='line', line_color='red', line_width=3, p=p)
p, shape_renderer2 = create_plot(x_2_values, y_2_values, plot_type='diamond', shape_fill_color='white', p=p)
p, line_renderer2 = create_plot(x_2_values, y_2_values, plot_type='line', line_color='green', line_width=3, p=p)

# Create the Legend
legend = Legend(items=[
    LegendItem(label='Circle Line', renderers=[shape_renderer1]),
    LegendItem(label='Red Line', renderers=[line_renderer1]),
    LegendItem(label='Diamond Line', renderers=[shape_renderer2]),
    LegendItem(label='Green Line', renderers=[line_renderer2])
])
p.add_layout(legend)

show(p)

# Perfect 3 
```
