```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool, ColumnDataSource

# Create some example data
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 7, 6]
labels = ['A', 'B', 'C', 'D', 'E']
common_label = "Data Points"

# Create a ColumnDataSource with the data
source = ColumnDataSource(data=dict(x=x, y=y, labels=labels))

# Create a figure
p = figure(width=1000, height=1000)

# Plot the data points
scatter = p.scatter('x', 'y', source=source, size=10)

# Define the tooltip
tooltips = [
    ("Title", common_label),
    ("Label", "@labels"),
    ("X", "@x"),
    ("Y", "@y"),
]

# Add the HoverTool with the defined tooltips to the figure
hover_tool = HoverTool(tooltips=tooltips, renderers=[scatter])
p.add_tools(hover_tool)

# Show the figure
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool, CustomJS
from bokeh.models import Div

# Create a figure
p = figure()

# Add some data points to the figure
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
circle_renderer = p.circle(x, y, size=10, color='blue')

# Create a custom tooltip
custom_tooltip = Div(width=200, height=50)

# Define a callback to update the custom tooltip
callback = CustomJS(args=dict(tooltip=custom_tooltip), code="""
    // Get the information about the hovered data point
    const x = cb_data['geometry']['x'];
    const y = cb_data['geometry']['y'];

    // Update the content of the custom tooltip
    tooltip.text = `Custom Tooltip<br>X: ${x.toFixed(2)}, Y: ${y.toFixed(2)}`;
""")

# Add the custom tooltip to the figure's tools
hover_tool = HoverTool(renderers=[circle_renderer], tooltips=None, callback=callback)
p.add_tools(hover_tool)

# Show the figure
show(p)
```
