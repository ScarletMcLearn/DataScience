**Import ploting library**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Distribution of Item Purchase**

```python
plt.figure(figsize=(20, 8))
df['Item Purchased'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Distribution of Item purchased')
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
item = df.groupby('Item Purchased')['Purchase Amount (USD)'].sum().reset_index()
item
```

**Item purchased by Amount(USD)**

```python
plt.figure(figsize=(12, 5))
sns.barplot(data=item,x='Item Purchased',y='Purchase Amount (USD)')
plt.title('Item purchased by Amount(USD)')
plt.ylabel('Purchase Amount (USD)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
