```python
'''
Here I defined a function `find_engine_type` that searches for an engine
type within a given title. It checks if any engine type from the `engine_types
` list is present in the title (case-insensitive).
If it finds a match, it returns the engine type; otherwise, it returns `None`.

Then, I applied this function to the `title` column of the `df` DataFrame
to create a new column called `parsed_engine_type`. This new column contains
the engine type found in each title. Finally, I displayed the updated
DataFrame.
'''

# Function to find and return the engine type in the title
def find_engine_type(title):
    for engine_type in engine_types:
        
        if engine_type.lower() in title.lower():
            return engine_type
    return None

# Apply the function to the 'title' column to create the 'parsed_engine_type' column
df['parsed_engine_type'] = df['title'].apply(find_engine_type)
df
```

```python
'''
Here I defined a function `clean_and_format` to clean and format
string values. The function performs the following steps:

1. **Remove extra spaces**: It replaces multiple spaces with a single
space and trims leading/trailing spaces.
2. **Add space between numbers and letters**: It ensures there is a
space between numbers and letters.
3. **Convert to title case**: It converts words to title case unless they
are in all capitals.

Then, I applied this function to all elements in the `df` DataFrame using
`applymap`, which applies the function to each element of the DataFrame.
Finally, I displayed the cleaned and formatted DataFrame.
'''

import re

def clean_and_format(value):
    if isinstance(value, str):
        # Remove extra spaces
        value = re.sub(r'\s+', ' ', value).strip()
        # Add space between numbers and letters
        value = re.sub(r'(\d)([a-zA-Z])', r'\1 \2', value)
        value = re.sub(r'([a-zA-Z])(\d)', r'\1 \2', value)
        # Convert to title case if not already in title case and not all capitals
        words = value.split()
        formatted_words = []
        for word in words:
            if word.isupper():
                formatted_words.append(word)
            else:
                formatted_words.append(word.title())
        value = ' '.join(formatted_words)
    return value

df = df.applymap(clean_and_format)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_violin_with_outliers(df, column_name, log=False):
    """
    Plots a violin plot with outliers for a specified column in a DataFrame.
    Optionally scales the x-axis logarithmically.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the data.
    column_name (str): The name of the column to plot.
    log (bool): Whether to scale the x-axis logarithmically.

    Returns:
    None
    """
    plt.figure(figsize=(10, 6))
    
    # Create a copy of the column to avoid modifying the original DataFrame
    data = df[column_name].copy()
    
    if log:
        data = np.log1p(data)
        xlabel = f'Log of {column_name}'
    else:
        xlabel = column_name

    sns.violinplot(x=data, inner="point", linewidth=1.5)
    plt.title(f'Violin Plot of {column_name}')
    plt.xlabel(xlabel)
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()

    
plot_violin_with_outliers(df, 'price_gross', log=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def plot_distribution(df, column_name, bins=10, log=False):
    """
    Plots a distribution plot for a specified column in a DataFrame.
    Optionally scales the x-axis logarithmically.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the data.
    column_name (str): The name of the column to plot.
    bins (int): The number of bins for the histogram.
    log (bool): Whether to scale the x-axis logarithmically.

    Returns:
    None
    """
    plt.figure(figsize=(10, 6))
    
    # Create a copy of the column to avoid modifying the original DataFrame
    data = df[column_name].copy()
    
    if log:
        data = np.log1p(data)
        xlabel = f'Log of {column_name}'
    else:
        xlabel = column_name

    sns.histplot(data, bins=bins, kde=True)
    plt.title(f'Distribution Plot of {column_name}')
    plt.xlabel(xlabel)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
    
plot_distribution(df, 'price_gross', bins=50, log=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

def plot_bar(df, column_name, log=False):
    """
    Plots a bar plot for a specified column in a DataFrame.
    Optionally scales the y-axis logarithmically and connects the tops of the bars with a smooth, navy line.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the data.
    column_name (str): The name of the column to plot.
    log (bool): Whether to scale the y-axis logarithmically.

    Returns:
    None
    """
    plt.figure(figsize=(10, 6))
    
    # Calculate the counts for each category
    value_counts = df[column_name].value_counts()
    
    # Create the bar plot
    ax = value_counts.plot(kind='bar', color='skyblue', log=log)
    
    # Plot a smooth line connecting the tops of the bars
    tops = value_counts.values
    x = np.arange(len(tops))
    
    try:
        x_smooth = np.linspace(x.min(), x.max(), 300)
        spl = make_interp_spline(x, tops, k=3)
        y_smooth = spl(x_smooth)
        ax.plot(x_smooth, y_smooth, color='navy', linewidth=2)
    except ValueError as e:
        print(f"Warning: {e}. Plotting without smoothing.")
        ax.plot(x, tops, marker='o', color='navy', linestyle='-', linewidth=2)
    
    # Set the title and labels
    plt.title(f'Bar Plot of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Log Frequency' if log else 'Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Tilt the x-axis labels
    plt.xticks(rotation=45, ha='right')
    
    # Show the plot
    plt.show()
    
plot_bar(df, 'color', log=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_average(df, group_by_column, value_column):
    # Group by the specified column and calculate the average of the value column
    average_price = df.groupby(group_by_column)[value_column].mean().reset_index()

    # Rename the columns for clarity
    average_price.columns = [group_by_column, f'average_{value_column}']

    # Create the scatter plot
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=average_price, x=group_by_column, y=f'average_{value_column}', color='blue', label=f'Average {value_column}')

    # Add a smooth line plot to show the trend
    sns.lineplot(data=average_price, x=group_by_column, y=f'average_{value_column}', color='aqua', label='Trend Line', errorbar=None)

    # Add titles and labels
    plt.title(f'Average {value_column} by {group_by_column.capitalize()}')
    plt.xlabel(group_by_column.capitalize())
    plt.ylabel(f'Average {value_column.capitalize()}')
    plt.legend()
    plt.xticks(rotation=45)

    # Show the plot
    plt.show()

plot_average(df, 'parsed_added_year', 'price_net')
```
