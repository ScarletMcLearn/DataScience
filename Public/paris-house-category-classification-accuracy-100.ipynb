```python
df['category'] = df['category'].replace({'Luxury': 1, 'Basic': 0})
df
```

```python
df = df.drop(columns=['cityCode', 'made'])
df
```

```python
df.columns
```

```python
col = ['squareMeters', 'numberOfRooms', 'hasYard', 'hasPool', 'floors',
       'cityPartRange', 'numPrevOwners', 'isNewBuilt', 'hasStormProtector',
       'basement', 'attic', 'garage', 'hasStorageRoom', 'hasGuestRoom',
       'price',]
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the DataFrame (excluding the 'category' column if it's your target)
df[col] = scaler.fit_transform(df[col])

df
```

```python
# Assuming your dataframe is named df and the target column is 'category'
X = df.drop('category', axis=1)
y = df['category']
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)
```

```python
import pandas as pd
from sklearn.utils import all_estimators
from sklearn.metrics import accuracy_score
from sklearn.exceptions import NotFittedError

# Initialize lists to store results
results = []

# Get all classifiers
estimators = all_estimators(type_filter='classifier')

# Iterate over all classifiers
for name, Classifier in estimators:
    try:
        # Initialize the classifier
        clf = Classifier()
        
        # Fit the classifier on the training data
        clf.fit(X_train, y_train)
        
        # Predict on the training data
        y_train_pred = clf.predict(X_train)
        
        # Predict on the test data
        y_test_pred = clf.predict(X_test)
        
        # Calculate accuracy
        train_accuracy = accuracy_score(y_train, y_train_pred)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        
        # Append results
        results.append({
            'Model': name,
            'Train Accuracy': train_accuracy,
            'Test Accuracy': test_accuracy
        })
    except (TypeError, ValueError, NotFittedError, IndexError):
        # Skip estimators that fail to fit or predict
        continue

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Sort the DataFrame by test accuracy
results_df = results_df.sort_values(by='Test Accuracy', ascending=False)


# Display the results
results_df
```
