# Get Abridged Text Data Of Several Of Shakespeare's Plays From Sparknotes

## Importing dependencies

```python
import bs4 as bs
import urllib.request 

import time
from random import randint

from os import linesep
```

```python
# Base Link
domain = "http://nfs.sparknotes.com/errors/page_"

#List to store scraped text
ModernTextList = [["List Of Modern Text"],["Speaker","Dialogue"]]
```

We go to the first page and last page of the base link to find the first page number and last page number.
Then we save it to 2 variables as our scraping parameters.

```python
starting_page = 2
finishing_page = 162
```

```python
def text_grabber(Tags, ModernTextList):
    for item in range(0, len(Tags)):

            for t in Tags[item]:
                if t.find("div", {"class": "modern-line"}):
                    try:
                        ModernTextList.append([str(t.find("b").text.replace("\n", " ")), str(t.find("div").text.replace("\n", " "))])
                    except:
                        continue
    pauser = randint(2, 5)
    print('Pausing for ' + str(pauser) + ' seconds.')
    time.sleep(pauser)
```

```python
def text_grabber_debugger(Tags, ModernTextList):
    for item in range(0, len(Tags)):

            for t in Tags[item]:
                if t.find("div", {"class": "modern-line"}):
                    try:
                        current_tag_data = [str(t.find("b").text.replace("\n", " ")), str(t.find("div").text.replace("\n", " "))]
                        ModernTextList.append(current_tag_data)
                        print("Current Page Data: " + linesep + str(current_tag_data))
                        
                    except Exception as e:
                        print(e)
                        pass
    pauser = randint(2, 5)
    print('Pausing for ' + str(pauser) + ' seconds.')
    time.sleep(pauser)
```

```python
for iter in range (starting_page, finishing_page+2, 2):
    link = domain + str(iter) + ".html"
    print("Now processing Link: " + link + linesep)
        
    sauce = urllib.request.urlopen(link).read() 
    soup = bs.BeautifulSoup(sauce, "lxml")
    Tags = soup.find_all("tr")
    
    text_grabber_debugger(Tags, ModernTextList)
    
    print(linesep + "Completed Link: "+ link + linesep)
```

Writing scraped text to file.

```python
def write_to_file(file_name, ModernTextList):
    file = open(file_name, 'w')

    for item in ModernTextList:
        file.write("%s\n" % item)

    file.close()
    print('Done')
```

```python
write_to_file('As_You_Like_It_Abridged_Text.txt', ModernTextList)
```
