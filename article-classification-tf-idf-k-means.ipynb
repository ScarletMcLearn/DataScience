```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def group_common_articles(df):
    """
    Groups common articles in a DataFrame based on cosine similarity of their titles.

    Args:
        df: pandas DataFrame with columns 'Article' (article body) and 'Title'.

    Returns:
        A list of lists, where each inner list contains the indices of articles 
        that are considered similar.
    """

    # 1. Create TF-IDF vectors from article titles
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(df['Title'])

    # 2. Calculate cosine similarity
    cosine_sim = cosine_similarity(tfidf_matrix)

    # 3. Define a threshold for similarity (adjust as needed)
    threshold = 0.7

    # 4. Group articles based on similarity
    groups = []
    for i in range(len(cosine_sim)):
        similar_articles = [j for j in range(len(cosine_sim)) if cosine_sim[i][j] > threshold]
        if len(similar_articles) > 1:
            groups.append(similar_articles)

    # 5. Remove duplicate groups
    unique_groups = []
    for group in groups:
        if group not in unique_groups:
            unique_groups.append(group)

    # 6. Print the results
    for group in unique_groups:
        print("Group:")
        for index in group:
            print(f"  - Title: {df['Title'][index]}")
            print(f"    Article: {df['Article'][index]}")
        print()

    return groups

# Example usage:
# df = pd.DataFrame({'Article': ['This is the first article.', 'This is the second article.', 
#                               'This is a similar article to the first one.', 'This is an unrelated article.'],
#                    'Title': ['First Article', 'Second Article', 'Similar Article', 'Unrelated Article']})

groups = group_common_articles(df)
groups
```
