```python
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

tmp_fle = tmp_fle.lower().strip().replace('\n', ' ').replace('\r', ' ')

tmp_file_no_punc = ""

for char in tmp_fle:
   if char not in punctuations:
       tmp_file_no_punc = tmp_file_no_punc + char
```

```python
import spacy
nlp = spacy.load('en')

def lemmatizer(text):        
    sent = []
    doc = nlp(text)
    for word in doc:
        sent.append(word.lemma_)
    return " ".join(sent)

lemm = lemmatizer(tmp_file_no_punc)
```

```python
def pos_neg_stat(doc):
    '''doc == str '''
    neg_occ = 0
    pos_occ = 0
    
    pos_ = []
    neg_ = []

    for _ in (doc.split()):
        if _ in pos:
            pos_occ = pos_occ + 1
            pos_.append(_)

        elif _ in neg:
            neg_occ = neg_occ + 1
            neg_.append(_)

    print("Positives: " + str(pos_occ) + '/' + str(len(doc.split())))
    print("Negatives: " + str(neg_occ) + '/' + str(len(doc.split())))
    
    doc_l = len(doc.split())
    pos_scr = len(pos_)/doc_l * 100
    neg_scr = len(neg_)/doc_l * 100
    
    print('Positive Percentage: ' + str(pos_scr))
    print('Negative Percentage: ' + str(neg_scr))
    
    return({'pos':pos_, 'neg':neg_, 'pos_scr':pos_scr, 'neg_scr':neg_scr, 'doc_len':doc_l})

sl = pos_neg_stat(cln_lemm)
```

```python
def m_emo_to_senti_percentage(posi, negi):
    if posi > negi:
        p = posi*100/(posi+negi)
        print('Positive: ' + str(p))
        return p
    n = negi*100/(posi+negi)
    print('Negative: ' + str(n))
    return n
```

```python
# import operator

# sorted_count = sorted(vectorizer.vocabulary_.items(), key=operator.itemgetter(0))

# # OrderedDict(vectorizer.vocabulary_.items())

def get_cmn_in_range(lower, upper, sorted_count):
    tmp = []
    
    for itm in sorted_count.most_common(upper):
        if itm not in sorted_count.most_common(lower):
            tmp.append(itm)
#             print(itm)
    return tmp
```
