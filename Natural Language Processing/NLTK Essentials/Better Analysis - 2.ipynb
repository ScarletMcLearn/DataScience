```python
postDocs = [x.lower() for x in docs]
```

```python
from nltk.corpus import stopwords

stopset = set(stopwords.words('english'))

stopset.update(['lt','p','/p','br','amp','quot','field','font','normal','span','0px','rgb','style','51', 
                'spacing','text','helvetica','size','family', 'space', 'arial', 'height', 'indent', 'letter'
                'line','none','sans','serif','transform','line','variant','weight','times', 'new','strong', 'video', 'title'
                'white','word','letter', 'roman','0pt','16','color','12','14','21', 'neue', 'apple', 'class',  ])
```

```python
from nltk.tokenize import word_tokenize 
from nltk.corpus import wordnet

for i in range(len(postDocs)):
    word_tokens = word_tokenize(postDocs[i])
    filtered_doc = [w for w in word_tokens if not w in stopset and wordnet.synsets(w)] 
    filtered_doc = ' '.join(filtered_doc)
    postDocs[i] = filtered_doc
    print("Done " + str(i))
    
```

```python
# import pandas and sklearn's CountVectorizer class
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# create a dataframe from a word matrix
def wm2df(wm, feat_names):
    
    # create an index for each row
    doc_names = ['Doc{:d}'.format(idx) for idx, _ in enumerate(wm)]
    df = pd.DataFrame(data=wm.toarray(), index=doc_names,
                      columns=feat_names)
    return(df)
  
# set of documents
corpora = postDocs

# instantiate the vectorizer object
cvec = CountVectorizer(lowercase=False)

# convert the documents into a document-term matrix
wm = cvec.fit_transform(corpora)

# retrieve the terms found in the corpora
tokens = cvec.get_feature_names()

# create a dataframe from the matrix
wm2df(wm, tokens)
```
