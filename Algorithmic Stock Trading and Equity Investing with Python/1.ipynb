```python
import pandas as pd

def transform_dataframe(input_df):
    # Pivot the dataframe to get the desired structure
    pivoted_df = input_df.pivot(index='date', columns='symbol', values=['close', 'ycp'])
    
    # Flatten the multi-level columns
    pivoted_df.columns = [' '.join(col).strip() for col in pivoted_df.columns.values]
    
    # Reset the index to make 'date' a column again
    pivoted_df.reset_index(inplace=True)
    
    # Rename columns to match the desired format
    pivoted_df.columns = ['date'] + [f'{col}_{agg}' for col in input_df['symbol'].unique() for agg in ['close', 'ycp']]
    
    # Count the number of unique symbols
    symbol_count = input_df.groupby('date')['symbol'].nunique().reset_index()
    
    # Merge the symbol count with the pivoted dataframe
    result_df = pd.merge(pivoted_df, symbol_count, on='date')
    
    # Rename the symbol count column
    result_df.rename(columns={'symbol': 'Number of Symbols'}, inplace=True)
    
    return result_df

data_1m = transform_dataframe(data_1m)
data_1m
```
